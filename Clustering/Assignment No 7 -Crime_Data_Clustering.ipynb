```python
crime['Clusters'] = yhc
```

```python
crime['Clusters'].value_counts()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['Clusters'],crime['Place'], c=yhc)
plt.show()
```

                     K-Means Clustering

```python
from sklearn.cluster import KMeans 
```

```python
data = pd.read_csv("crime_data.csv")
data
```

```python
#Normalize Data
norm = MinMaxScaler()
newdf = norm.fit_transform(data.iloc[:,1:].to_numpy())
norm_df = pd.DataFrame(newdf, columns=data.columns[1:])
norm_df
```

```python
# Elbow curve
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(norm_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curve')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above  plot
model=KMeans(n_clusters=4)
model.fit(norm_df)
model.labels_
```

```python
x=pd.Series(model.labels_)
data['Clusters']=x
```

```python
data
```

```python
data.iloc[:,1:5].groupby(data.Clusters).mean()
```

```python
data.groupby(['Clusters'])['Unnamed: 0'].apply(lambda x : tuple(x)) 
```

                           DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("crime_data.csv")
df
```

```python
df.boxplot() 
```

```python
df = df.rename({'Unnamed: 0':'Place'},axis = 1)
df
```

```python
array = df.iloc[:,1:].values
```

```python
array
```

```python
stscaler = StandardScaler()
X = stscaler.fit_transform(array)  #standardization
X
```

```python
from sklearn.metrics import silhouette_score

eps_values = [0.8,1,1.1,1.2,1.3,1.4,1.5,1.6]
min_samples_values = [1,2,3,4]

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan_df = DBSCAN(eps=eps,min_samples=min_samples)
        dbscan_df.fit(X)
        labels = dbscan_df.labels_
        print("For eps :",eps,"and min_sample:",min_samples," Sil_score is:",silhouette_score(df.iloc[:,1:], labels))
    continue
```

```python
#from above result we can say that min_sample =4 & eps = 1.1 will be good choise.
dbscan = DBSCAN(eps = 1.1, min_samples = 4)  
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
df['Clusters'] = dbscan.labels_
df
```

```python
df[df['Clusters']== -1]
```

```python
df['Clusters'].value_counts() 
```

```python
df.groupby(df['Clusters']).mean()
```

```python
import matplotlib.pyplot as plt

# Plot Clusters
plt.figure(figsize=(12, 8))  
plt.scatter(df['Clusters'],df['Place'], c=dbscan.labels_)
plt.show()
```

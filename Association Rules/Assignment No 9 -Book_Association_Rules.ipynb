```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
book = pd.read_csv("book.csv")
book
```

```python
book.shape
```

```python
book.describe()
```

```python
book.info()
```

                               Aprori Algorithm

```python
frequent_itemset1 = apriori(book, min_support = 0.1, use_colnames=True,verbose =1)
frequent_itemset1
```

```python
frequent_itemset1.shape
```

```python
frequent_itemset1.sort_values("support", ascending = False).head()
```

```python
rules1 = association_rules(frequent_itemset1, metric ='confidence', min_threshold = 0.3)
rules1
```

```python
rules1[(rules1.support > 0.015) & (rules1.confidence > 0.4)].sort_values("confidence",
                                                                         ascending = False).shape
```

```python
rules1['lhs items'] = rules1['antecedents'].apply(lambda x:len(x) )
rules1[rules1['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
rules1['antecedents_'] = rules1['antecedents'].apply(lambda a: ','.join(list(a)))
rules1['consequents_'] = rules1['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules1[rules1['lhs items']>1].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'confidence')
# Generate a heatmap with annotations 
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

```python
rules1[['support', 'confidence', 'lift']].hist()
plt.show()
```

                                 FpGrowth Algorithm

```python
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets_fp=fpgrowth(book, min_support=0.13, use_colnames=True,verbose=1)
print(frequent_itemsets_fp.shape)
```

```python
frequent_itemsets_fp.sort_values("support", ascending = False).head()
```

```python
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.5)
print(rules_fp.shape)
```

```python
rules_fp[(rules_fp.support > 0.15) & (rules_fp.confidence > 0.5)].sort_values("confidence",
                                                                              ascending = False).head()
```

```python
rules_fp['lhs items'] = rules_fp['antecedents'].apply(lambda x:len(x) )
rules_fp[rules_fp['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
rules_fp['antecedents_'] = rules_fp['antecedents'].apply(lambda a: ','.join(list(a)))
rules_fp['consequents_'] = rules_fp['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules_fp[rules_fp['lhs items']>1].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'confidence')
# Generate a heatmap with annotations 
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
sns.scatterplot( x = rules_fp['support'], y = rules_fp['confidence'])
```

```python
rules_fp_li = association_rules(frequent_itemsets_fp, metric="lift", min_threshold=0.6)
print(rules_fp_li.shape)
```

```python
rules_fp_li['lhs items'] = rules_fp_li['antecedents'].apply(lambda x:len(x) )
rules_fp_li[rules_fp_li['lhs items']>1].sort_values('lift', ascending=False).head()
```
